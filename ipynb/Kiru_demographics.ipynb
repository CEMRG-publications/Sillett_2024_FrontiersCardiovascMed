```python
# Define a function to calculate the new column values based on another column
def postprocess_HF_meds(entry):
    # Perform calculations based on the entry and return the result
    
    string = entry.replace(" ", "")
    ans = string.split(",")
    
    return ans
```

```python
HF_meds = af_df['HF meds'].apply(postprocess_HF_meds)
```

```python
HF_meds
```

```python
my_list = []

for i in range(0, len(HF_meds)):
    my_list += HF_meds.iloc[i]
```

```python
np.unique(my_list)
```

```python
def count_instances(lst, value):
    count = 0
    for element in lst:
        if element == value:
            count += 1
    return count
```

```python
for word in np.unique(my_list):
    
    ans = count_instances(my_list, word)
    
    print(f"{word}\t\t number of instances: \t\t", ans)
```
