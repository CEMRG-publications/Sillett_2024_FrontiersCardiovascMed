```python
import pandas as pd
import numpy as np
```

```python
def Test_ReRead(Case, Res, Cutter):

    filename = f'{Res}-{Cutter}-AbsoluteError.csv'

    test_reread = pd.read_csv(f'/home/csi20local/Data/RG_CT_Cases/CT-CRT-{Case}/{filename}', 
                              sep = ' ', index_col = 0)
    
    return test_reread
```

```python
"""
Create CSV Absolute Local errors files for FIMH 10 Cases
"""

def CreateCSVFiles(Res, Cutter):

    for Case in ['01', '02', '05', '06', '07', '08', '12', '15', '16']:

        pC_distances_dflist = []
        
        Cutter_to_Number_dict = {
          "LAA": 0,
          "LSPV": 1,
          "LIPV": 2,
          "RSPV": 3,
          "RIPV": 4
        }

        CutterNum = Cutter_to_Number_dict[Cutter]

        Path = f'/home/csi20local/Data/RG_CT_Cases/CT-CRT-{Case}/{Res}'

        for SW in ['9e-2', '3e-2', '9e-3', '3e-3', '9e-4', '3e-4', '0.0']:
            for BE in ['4e-6', '1e-6', '4e-7', '1e-7', '4e-8', '1e-8', '4e-9', '1e-9', '4e-10']:
                df = pd.read_csv(f'{Path}/SW-{SW}-BE-{BE}/pC-distances.txt', sep=' ', names=['Error'])

                pC_distances_dflist.append(df)

        array = np.zeros(shape=(9,7))

        for j in range(0,7):
            for i in range(0,9):
                array[i][j] = pC_distances_dflist[9*j+i].loc[CutterNum][0]

        array_df = pd.DataFrame(array, columns=['9e-2', '3e-2', '9e-3', '3e-3', '9e-4', '3e-4', '0.0'],
                     index=['4e-6', '1e-6', '4e-7', '1e-7', '4e-8', '1e-8', '4e-9', '1e-9', '4e-10'])

        filename = f'{Res}-{Cutter}-AbsoluteError.csv'

        array_df.to_csv(f'/home/csi20local/Data/RG_CT_Cases/CT-CRT-{Case}/{filename}', sep=' ')

        test_reread = pd.read_csv(f'/home/csi20local/Data/RG_CT_Cases/CT-CRT-{Case}/{filename}', 
                   sep = ' ', index_col = 0)
```

### Use this Function for Checking CSVs created:

```python
def CHECK_CSV_Creation_AbsoluteError(Res, Cutter, Case):
    
    Cutter_to_Number_dict = {
          "LAA": 0,
          "LSPV": 1,
          "LIPV": 2,
          "RSPV": 3,
          "RIPV": 4
        }

    CutterNum = Cutter_to_Number_dict[Cutter]

    PV_csv = pd.read_csv(f'/home/csi20local/Data/RG_CT_Cases/{Case}/{Res}-{Cutter}-AbsoluteError.csv', 
                                  sep = ' ', index_col = 0)

    status = True

    for BE in ['4e-6', '1e-6', '4e-7', '1e-7', '4e-8', '1e-8', '4e-9', '1e-9', '4e-10']:
            for SW in ['9e-2', '3e-2', '9e-3', '3e-3', '9e-4', '3e-4', '0.0']:
                PV_txt = pd.read_csv(f"/home/csi20local/Data/RG_CT_Cases/{Case}/{Res}/SW-{SW}-BE-{BE}/pC{CutterNum}-distance.txt",
                                    names=["Error"])
                PV_txt = PV_txt.iloc[0][0]


                if (round(float(PV_txt), 6) == round(PV_csv[f'{SW}'][float(BE)], 6)):
                    status = True

                else:
                    print(f"Problem with Case {Case}, Error at BE {BE} SW {SW}")
                    status=False
                    break
                    
    return print(status)
```
