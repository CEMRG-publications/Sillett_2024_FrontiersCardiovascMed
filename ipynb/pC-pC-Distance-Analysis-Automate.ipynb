```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

## Inter PV-PV Distance Percentage Error

```python
Case = '07'
```

```python
Path = f'/home/csi20local/Data/RG_CT_Cases/CT-CRT-{Case}/MT'
Path_HR = f'/home/csi20local/Data/RG_CT_Cases/CT-CRT-{Case}/MT-HiRes'

## Read in MT dataframes for PV-PV distance percentage errors

## Each SW BE combination has its own df of PV-PV distances

SW_BE_df_list = []

for SW in ['9e-2', '3e-2', '9e-3', '3e-3', '9e-4', '3e-4', '0.0']:
    for BE in ['4e-6', '1e-6', '4e-7', '1e-7', '4e-8', '1e-8', '4e-9', '1e-9', '4e-10']:
        
        SW_BE_df_list.append(pd.read_csv(f'{Path}/SW-{SW}-BE-{BE}/pC-pC-distances.txt', sep=' ', names=['Error']))
        
SW_BE_popASD = pd.read_csv(f'{Path}/SW-3e-3-BE-1e-6/pC-pC-distances.txt', sep=' ', names=['Error'])
        
## Read in MT-HiRes dataframes for PV-PV distance percentage errors

SW_BE_df_list_HiRes = []

for SW in ['9e-2', '3e-2', '9e-3', '3e-3', '9e-4', '3e-4', '0.0']:
    for BE in ['4e-6', '1e-6', '4e-7', '1e-7', '4e-8', '1e-8', '4e-9', '1e-9', '4e-10']:
        
        SW_BE_df_list_HiRes.append(pd.read_csv(f'{Path_HR}/SW-{SW}-BE-{BE}/pC-pC-distances.txt', sep=' ', names=['Error']))

SW_BE_popASD_HiRes = pd.read_csv(f'{Path_HR}/SW-9e-3-BE-1e-6/pC-pC-distances.txt', sep=' ', names=['Error'])
        
## Create lists of PV-PV distance errors across all SW BE combinations for MT

pC0_pC1 = []
pC0_pC2 = []
pC0_pC3 = []
pC0_pC4 = []

pC1_pC2 = []
pC1_pC3 = []
pC1_pC4 = []

pC2_pC3 = []
pC2_pC4 = []

pC3_pC4 = []

pop_errors = []

for i in range(0,63):
        
    reshaped_array = np.reshape(SW_BE_df_list[i].to_numpy(), (5,5))
    
    reshaped_array = np.abs(reshaped_array)
    
    pC_df = pd.DataFrame(data=reshaped_array, columns=['pC0', 'pC1', 'pC2', 'pC3', 'pC4'],
                         index=['pC0', 'pC1', 'pC2', 'pC3', 'pC4'])
    
    pC0_pC1.append(pC_df.loc['pC0']['pC1'])
    pC0_pC2.append(pC_df.loc['pC0']['pC2'])
    pC0_pC3.append(pC_df.loc['pC0']['pC3'])
    pC0_pC4.append(pC_df.loc['pC0']['pC4'])
    
    pC1_pC2.append(pC_df.loc['pC1']['pC2'])
    pC1_pC3.append(pC_df.loc['pC1']['pC3'])
    pC1_pC4.append(pC_df.loc['pC1']['pC4'])
    
    pC2_pC3.append(pC_df.loc['pC2']['pC3'])
    pC2_pC4.append(pC_df.loc['pC2']['pC4'])
    
    pC3_pC4.append(pC_df.loc['pC3']['pC4'])
    
reshape = np.reshape(SW_BE_popASD.to_numpy(), (5,5))
reshape = np.abs(reshape)
pop_df = pd.DataFrame(data=reshape, columns=['pC0', 'pC1', 'pC2', 'pC3', 'pC4'],
                         index=['pC0', 'pC1', 'pC2', 'pC3', 'pC4'])

pop_errors.append(pop_df.loc['pC0']['pC1'])
pop_errors.append(pop_df.loc['pC0']['pC2'])
pop_errors.append(pop_df.loc['pC0']['pC3'])
pop_errors.append(pop_df.loc['pC0']['pC4'])
    
pop_errors.append(pop_df.loc['pC1']['pC2'])
pop_errors.append(pop_df.loc['pC1']['pC3'])
pop_errors.append(pop_df.loc['pC1']['pC4'])
    
pop_errors.append(pop_df.loc['pC2']['pC3'])
pop_errors.append(pop_df.loc['pC2']['pC4'])
    
pop_errors.append(pop_df.loc['pC3']['pC4'])
    
## MT-HiRes

pC0_pC1_ = []
pC0_pC2_ = []
pC0_pC3_ = []
pC0_pC4_ = []

pC1_pC2_ = []
pC1_pC3_ = []
pC1_pC4_ = []

pC2_pC3_ = []
pC2_pC4_ = []

pC3_pC4_ = []

pop_errors_ = []

for i in range(0,63):
        
    reshaped_array = np.reshape(SW_BE_df_list_HiRes[i].to_numpy(), (5,5))
    
    reshaped_array = np.abs(reshaped_array)
    
    pC_df = pd.DataFrame(data=reshaped_array, columns=['pC0', 'pC1', 'pC2', 'pC3', 'pC4'],
                         index=['pC0', 'pC1', 'pC2', 'pC3', 'pC4'])
    
    pC0_pC1_.append(pC_df.loc['pC0']['pC1'])
    pC0_pC2_.append(pC_df.loc['pC0']['pC2'])
    pC0_pC3_.append(pC_df.loc['pC0']['pC3'])
    pC0_pC4_.append(pC_df.loc['pC0']['pC4'])
    
    pC1_pC2_.append(pC_df.loc['pC1']['pC2'])
    pC1_pC3_.append(pC_df.loc['pC1']['pC3'])
    pC1_pC4_.append(pC_df.loc['pC1']['pC4'])
    
    pC2_pC3_.append(pC_df.loc['pC2']['pC3'])
    pC2_pC4_.append(pC_df.loc['pC2']['pC4'])
    
    pC3_pC4_.append(pC_df.loc['pC3']['pC4'])
    
reshape_ = np.reshape(SW_BE_popASD_HiRes.to_numpy(), (5,5))
reshape_ = np.abs(reshape_)
pop_df_ = pd.DataFrame(data=reshape_, columns=['pC0', 'pC1', 'pC2', 'pC3', 'pC4'],
                         index=['pC0', 'pC1', 'pC2', 'pC3', 'pC4'])

pop_errors_.append(pop_df_.loc['pC0']['pC1'])
pop_errors_.append(pop_df_.loc['pC0']['pC2'])
pop_errors_.append(pop_df_.loc['pC0']['pC3'])
pop_errors_.append(pop_df_.loc['pC0']['pC4'])
    
pop_errors_.append(pop_df_.loc['pC1']['pC2'])
pop_errors_.append(pop_df_.loc['pC1']['pC3'])
pop_errors_.append(pop_df_.loc['pC1']['pC4'])
    
pop_errors_.append(pop_df_.loc['pC2']['pC3'])
pop_errors_.append(pop_df_.loc['pC2']['pC4'])
    
pop_errors_.append(pop_df_.loc['pC3']['pC4'])
    
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(20,5), dpi=150, sharey=True)

ax1.boxplot(x=[pC0_pC1, pC0_pC2, pC0_pC3, pC0_pC4,
              pC1_pC2, pC1_pC3, pC1_pC4,
              pC2_pC3, pC2_pC4, pC3_pC4], medianprops=dict(c='black'))

ax2.boxplot(x=[pC0_pC1_, pC0_pC2_, pC0_pC3_, pC0_pC4_,
              pC1_pC2_, pC1_pC3_, pC1_pC4_,
              pC2_pC3_, pC2_pC4_, pC3_pC4_], medianprops=dict(c='black'))

axes_labels = ['LAA-LSPV', 'LAA-LIPV', 'LAA-RSPV', 'LAA-RIPV',
              'LSPV-LIPV', 'LSPV-RSPV', 'LSPV-RIPV',
              'LIPV-RSPV', 'LIPV-RIPV',
              'RSPV-RIPV']

ax1.scatter(list(np.arange(1,11,1)), pop_errors)
ax2.scatter(list(np.arange(1,11,1)), pop_errors_)

# ax1.set_xlabel(axes_labels)
x = range(1,11)
ax1.set_xticks(x)
ax1.set_xticklabels(axes_labels, rotation=45)
ax1.set_ylabel('Percentage Error', fontsize=15)
# ax1.set_ylim(0, 14)
ax1.tick_params(labelsize=15)

ax2.set_xticks(x)
ax2.set_xticklabels(axes_labels, rotation=45)
ax2.tick_params(labelsize=15)

ax1.set_title('MT', fontsize=15)
ax2.set_title('MT-HiRes', fontsize=15)

# ax2.xticks(fontsize=20)

plt.suptitle(f'Case CT-CRT-{Case} PV to PV distance errors', fontsize=20)
```

## Clipper Position Absolute Errors [mm]

```python
Case = '02'
```

```python
Path = f'/home/csi20local/Data/RG_CT_Cases/CT-CRT-{Case}/MT'
Path_HR = f'/home/csi20local/Data/RG_CT_Cases/CT-CRT-{Case}/MT-HiRes'

## Construct dataframes for absolute distance errors [mm] for MT
## Each SW BE combination will give a 1-D Dataframe of position errors for all PVs.

df_list = []
pop = []

for SW in ['9e-2', '3e-2', '9e-3', '3e-3', '9e-4', '3e-4', '0.0']:
    for BE in ['4e-6', '1e-6', '4e-7', '1e-7', '4e-8', '1e-8', '4e-9', '1e-9', '4e-10']:
        
        df_list.append(pd.read_csv(f'{Path}/SW-{SW}-BE-{BE}/pC-distances.txt', sep=' ', names=['Error']))

pop = pd.read_csv(f'{Path}/SW-3e-3-BE-1e-6/pC-distances.txt', sep=' ', names=['Error'])
        
df_list_HR = []
pop_HR = []

for SW in ['9e-2', '3e-2', '9e-3', '3e-3', '9e-4', '3e-4', '0.0']:
    for BE in ['4e-6', '1e-6', '4e-7', '1e-7', '4e-8', '1e-8', '4e-9', '1e-9', '4e-10']:
        
        df_list_HR.append(pd.read_csv(f'{Path_HR}/SW-{SW}-BE-{BE}/pC-distances.txt', sep=' ', names=['Error']))
        
pop_HR = pd.read_csv(f'{Path_HR}/SW-9e-3-BE-1e-6/pC-distances.txt', sep=' ', names=['Error'])
        
## Construct lists of PV distance errors for each PV across all SW BE combinations

pC0 = []
pC1 = []
pC2 = []
pC3 = []
pC4 = []
pop_errors = []

for i in range(0,63):
        
    df = df_list[i]
    
    pC0.append(df.loc[0][0])
    pC1.append(df.loc[1][0])
    pC2.append(df.loc[2][0])
    pC3.append(df.loc[3][0])
    pC4.append(df.loc[4][0])
    
pop_errors.append(pop.loc[0][0])
pop_errors.append(pop.loc[1][0])
pop_errors.append(pop.loc[2][0])
pop_errors.append(pop.loc[3][0])
pop_errors.append(pop.loc[4][0])
    
pC0_ = []
pC1_ = []
pC2_ = []
pC3_ = []
pC4_ = []
pop_errors_ = []

for i in range(0,63):
        
    df = df_list_HR[i]
    
    pC0_.append(df.loc[0][0])
    pC1_.append(df.loc[1][0])
    pC2_.append(df.loc[2][0])
    pC3_.append(df.loc[3][0])
    pC4_.append(df.loc[4][0])
    
pop_errors_.append(pop_HR.loc[0][0])
pop_errors_.append(pop_HR.loc[1][0])
pop_errors_.append(pop_HR.loc[2][0])
pop_errors_.append(pop_HR.loc[3][0])
pop_errors_.append(pop_HR.loc[4][0])
    
axes_labels_abs = ['LAA', 'LSPV', 'LIPV', 'RSPV', 'RIPV']

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(20,5), dpi=150, sharey=True)

ax1.boxplot(x=[pC0, pC1, pC2, pC3, pC4], medianprops=dict(c='black'))

ax2.boxplot(x=[pC0_, pC1_, pC2_, pC3_, pC4_], medianprops=dict(c='black'))

ax1.scatter(list(np.arange(1,6,1)),pop_errors, s=100)
ax2.scatter(list(np.arange(1,6,1)),pop_errors_, s=100)

# ax1.set_xlabel(axes_labels)
x = range(1,6)
ax1.set_xticks(x)
ax1.set_xticklabels(axes_labels_abs, rotation=45)
ax1.set_ylabel('Distance [mm]', fontsize=15)
# ax1.set_ylim(0, 8)
ax1.tick_params(labelsize=15)

ax2.set_xticks(x)
ax2.set_xticklabels(axes_labels_abs, rotation=45)
ax2.tick_params(labelsize=15)

ax1.set_title('MT', fontsize=15)
ax2.set_title('MT-HiRes', fontsize=15)

# ax2.xticks(fontsize=20)

plt.suptitle(f'Case CT-CRT-{Case} PV CoM Position Errors', fontsize=20)

# Note

# Boxplots of MT are expected to be more bunched together than MT-HiRes and have smaller range since  the 
# resolution of the registration is higher, and so expect more deformation of PV discs?
```
