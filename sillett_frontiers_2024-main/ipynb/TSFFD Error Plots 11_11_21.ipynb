```python
ASD_opt = Retrieve_data('ASD', 'MT-HiRes', '0.0', 4e-9)
ASD_base = Retrieve_data('ASD', 'MT', '9e-3', 4e-7)

DHD_base = Retrieve_data('DHD', 'MT', '9e-3', 4e-7)
DHD_opt = Retrieve_data('DHD', 'MT-HiRes', '0.0', 4e-9)

DSC_base = Retrieve_data('DSC', 'MT', '9e-3', 4e-7)
DSC_opt = Retrieve_data('DSC', 'MT-HiRes', '0.0', 4e-9)
```

```python
fig, ((ax1, ax2, ax3)) = plt.subplots(1, 3, figsize=(14,4), dpi=150)

sns.set_style("whitegrid")

sns.boxplot(data=[ASD_base, ASD_opt], ax=ax1)
sns.swarmplot(data=[ASD_base, ASD_opt], color='black', ax=ax1)
ax1.set_ylabel('ASD [mm]', fontsize=15)

sns.boxplot(data=[DHD_base, DHD_opt], ax=ax2)
sns.swarmplot(data=[DHD_base, DHD_opt], color='black', ax=ax2)
ax2.set_ylabel('DHD [mm]', fontsize=15)

sns.boxplot(data=[DSC_base, DSC_opt], ax=ax3)
sns.swarmplot(data=[DSC_base, DSC_opt], color='black', ax=ax3)
ax3.set_ylabel('DSC', fontsize=15)

# statistical annotation
x1, x2 = 0, 1   
y = 1.8
h = 0.05
col = 'black'

ax1.text((x1+x2)*.5, y+h+0.02, "p value=6.57e-07", ha='center', va='bottom')
ax1.plot([x1, x1, x2, x2], [y, y+h, y+h, y], lw=1.5, c=col)
ax1.set_ylim(0, 2.0)
ax1.set_xticks([0,1])
ax1.set_xticklabels(['Baseline', 'Optimised'])
ax1.set_xlabel('Configuration')

y2 = 16
h2 = 0.5

ax2.text((x1+x2)*.5, y2+h2+0.02, "p value=0.17", ha='center', va='bottom')
ax2.plot([x1, x1, x2, x2], [y2, y2+h2, y2+h2, y2], lw=1.5, c=col)
ax2.set_ylim(0, 18.0)
ax2.set_xticks([0,1])
ax2.set_xticklabels(['Baseline', 'Optimised'])
ax2.set_xlabel('Configuration')

y3 = 1.0
h3 = 0.005

ax3.text((x1+x2)*.5, y3+h3+0.002, "p value=0.027", ha='center', va='bottom')
ax3.plot([x1, x1, x2, x2], [y3, y3+h3, y3+h3, y3], lw=1.5, c=col)
ax3.set_ylim(0.68, 1.03)
ax3.set_xticks([0,1])
ax3.set_xticklabels(['Baseline', 'Optimised'])
ax3.set_xlabel('Configuration')

fig.tight_layout() 
```
